```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
gorjeta = sns.load_dataset("tips")
gorjeta.head()
```

```python
sns.relplot(x='total_bill', y='tip', hue='sex', data=gorjeta)
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='day',style='time', data=gorjeta)
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', kind='swarm', data=gorjeta)
plt.show()
```

```python
sns.lmplot(x='total_bill',
           y='tip', 
           hue='smoker', 
           col='time', 
           palette='Set1', 
           data=gorjeta, 
           scatter_kws={'s':90, 'alpha': 0.5}, 
           line_kws={'lw': 2}
           )
plt.show()
```

```python
df_cereal = pd.read_csv('Dados/cereal.csv')
```

```python
df_cereal.head()
```

```python
df_cereal['calories per protein'] = df_cereal['calories'] / df_cereal['protein']
df_cereal.head()
```

```python
sns.scatterplot(x='calories per protein',
                y='name',
                hue='mfr',
                data=df_cereal
)
plt.show()
```

```python
sns.catplot(x='protein',
            y='calories',
            hue='mfr',
            data=df_cereal
            )
plt.show()
```

```python
df_cereal_abaixo_30 = df_cereal[df_cereal['rating'] < 30][['name','rating']]
df_cereal_acima_70 = df_cereal[df_cereal['rating'] > 70][['name','rating']]

print(f'Cereals with rating below 30:\n {df_cereal_abaixo_30}')
print(f'Cereals with rating above 70:\n {df_cereal_acima_70}')
```
