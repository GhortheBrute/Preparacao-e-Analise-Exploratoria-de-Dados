```python
import pandas as pd
import io
import matplotlib.pyplot as plt
```

```python
# Read and preprocess the file into a string
with open('Dados\\56320Desafio-tabela-Anliseunivariadadedados.csv', 'r', encoding='utf-8') as infile:
    lines = []
    for line in infile:
        stripped = line.rstrip('\n')
        if stripped.startswith('"') and stripped.endswith('"'):
            stripped = stripped[1:-1]
        stripped = stripped.replace('""', '"')
        lines.append(stripped)
    csv_string = '\n'.join(lines)

# Use StringIO to create an in-memory file
csv_buffer = io.StringIO(csv_string)

# Now read with pandas as if it were a file
df = pd.read_csv(csv_buffer, sep=',', encoding='utf-8')
```

```python
pais_counts = df['pais'].value_counts()

plt.figure(figsize=(10, 6))
pais_counts.plot(kind='bar', color='skyblue')
plt.title('Contagem de vinho por País')
plt.xlabel('País')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
#plt.tight_layout()
plt.show()
```

```python
pais_counts[::-1].plot(kind='barh')
#plt.tight_layout()
plt.show()
```

```python
vinho_por_pontuacao = df['pontos'].value_counts().sort_index(ascending=False)

vinho_por_pontuacao.plot(kind='bar', color='lightgreen')
plt.show()
```

```python
#distribuicao_de_precos = df['preco'].std()

df['preco'].plot(kind='hist', bins=50, color='lightgreen')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.title('Histograma dos preços')
plt.tight_layout()
plt.show()
```
